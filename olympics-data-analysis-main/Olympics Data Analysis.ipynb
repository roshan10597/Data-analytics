# Olympics Data Analysis


```python
import pandas as pd
import numpy as np
```

```python
#Import data
df=pd.read_csv("athlete_events.csv")
region_df=pd.read_csv("noc_regions.csv")
```

```python
df.tail()
```

```python
df.shape
```

```python
# Keep the data of summer olympics
df=df[df["Season"]=="Summer"]
```

```python
df.shape
```

```python
region_df.head()
```

```python
#left join on both dataframes
df=df.merge(region_df,on="NOC",how="left")
```

```python
df.head()
```

```python
df["region"].unique()
```

```python
df["region"].unique().shape[0]
```

```python
#Check for null values
df.isnull().sum()
```

```python
#Check for duplicate values
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['Medal'].value_counts()
```

```python
# One hot enocde the Medal column
pd.get_dummies(df["Medal"])
```

```python
#Merge main and one hot encode table
df=pd.concat([df,pd.get_dummies(df["Medal"])],axis=1)
```

```python
df.head()
```

```python

df.groupby("NOC").sum()[["Gold","Silver","Bronze"]].sort_values("Gold",ascending=False).reset_index()
```

```python
# above query returning sum of medals which is way more than the actual therfore to analyze the problem lets check for india
```

```python
df[(df["NOC"]=="IND") & (df["Medal"]=="Gold")]
```

```python
#As a team sport consist many players and in our data each medal of those players is also count so we have to remove it
df.drop_duplicates(subset=["Team","NOC","Games","Year","City","Sport","Event","Medal"])
```

```python
medal_tally=df.drop_duplicates(subset=["Team","NOC","Games","Year","City","Sport","Event","Medal"])
```

```python
medal_tally.shape
```

```python
groupby_medal_tally=medal_tally.groupby("NOC").sum()[["Gold","Silver","Bronze"]].sort_values("Gold",ascending=False).reset_index()
```

```python
groupby_medal_tally[(groupby_medal_tally["NOC"]=="IND")]
```

```python
medal_tally.to_csv("Olympics_Data.csv",sep=",")
```
